```python
def sumMatrixWhile(A):
    # Using a while loop, return the sum of all element of matrix A
    s = 0
    i = 0
    while i < A.shape[0] :
        j = 0
        while j < A.shape[1] :
            s = s + A[i,j]
            j = j + 1
        i = i + 1
    
    return s
```

```python
def initParamsANN(X, y, hidden_units):
    X = X.T
    y = y.T
    # number of inputs
    n = X.shape[0]
    # number of outputs
    n_y = y.shape[0]
    
    w1 = np.random.rand( hidden_units , n ) - 0.5
    b1 = np.random.rand( hidden_units , 1 ) - 0.5
    w2 = np.random.rand( n_y , hidden_units ) - 0.5
    b2 = np.random.rand( n_y , 1 ) - 0.5
    
    return w1, b1, w2, b2
```
